```python

fs = 50
time_duration = 30
serial_port = "COM8" 
baud_rate = 115200 
```

```python
import scipy.signal as sgnl
```

```python
[b, a] = sgnl.butter(1, [(0.5*2)/fs, (4 * 2)/fs], 'band')
```

```python
def preprocess(data_buffer, fs=50):
    
    x = np.arange(len(data_buffer))
    trend = np.polyfit(x, data_buffer, 1)[0] * x
    detrended_data = data_buffer - trend

    normalized_data = (detrended_data - np.mean(detrended_data)) / np.std(detrended_data)

    return normalized_data
```

```python
ser = serial.Serial(serial_port, baud_rate, timeout=1)
data_buffer = deque(maxlen=fs * 10)  # Circular buffer for efficient plotting

# Configure plot
plt.ion()
fig, ax = plt.subplots(figsize=(10, 6))

line, = ax.plot([], [], lw=2)
line2, = ax.plot([], [], lw=2, marker='o', color='r')

ax.set_ylim(-3, 3)
ax.set_xlim(0, int(fs*7))  


zi = sgnl.lfilter_zi(b, a)

find_peaks_params = {'height': 0.5, 'distance': fs // 2}

start = timer()
for _ in range(time_duration * fs):
    try:
        line_data = ser.readline().decode("utf-8").strip()
        if not line_data:
            continue
        
        ir_value = float(line_data.split(",")[0].split(":")[1])
        #ir_value = float(values[0].split(":")[1])

        data_buffer.append(ir_value)

        if len(data_buffer) < (fs * 7):
            data_buffer_2 = np.array(data_buffer)
        else:
            data_buffer_2 = np.array(data_buffer)
            data_buffer_2, zi_new = sgnl.lfilter(b, a, data_buffer_2, zi=zi*data_buffer_2[0])[-(fs * 7):]

        data_buffer_2 = preprocess(data_buffer_2)
        peaks, _ = sgnl.find_peaks(data_buffer_2, **find_peaks_params)
        line.set_data(range(len(data_buffer_2)), data_buffer_2)
        line2.set_data(peaks, data_buffer_2[peaks])

        # if len(peaks) > 1:
        #     rr_intervals = np.diff(peaks) / fs
        #     heart_rate = 60 / np.mean(rr_intervals)
        #     print(f"Heart rate: {heart_rate:.2f} BPM")
        
        plt.pause(0.001)
    except (ValueError, IndexError):
        ser.close
        continue

end = timer()
print(f"Data acquisition completed in {end - start:.2f} seconds.")
ser.close()
plt.ioff()
plt.show()
```
